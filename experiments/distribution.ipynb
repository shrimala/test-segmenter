```python
import pandas as pd
import plotly.express as px
import os
os.chdir("../")
!pwd
```

```python
from glob import glob
files = glob("./experiments/books/**/*.txt")
for file in files:
    break
```

```python
def get_paragraphs(file):
    with open(file, "r") as f:
        out = f.read().replace("\n========", "\n\n").split("\n\n")

    output = [x.replace("\n", "") for x in out if len(x.split())>10]
    return output
```

```python
print(" files :", len(files))
```

```python
output = []
for file in files:
    output +=get_paragraphs(file)
```

```python
word_count = [len(x.split(" ")) for x in output]
```

```python
df = pd.DataFrame({"paragraph_word_count":word_count})
```

```python
fig = px.histogram(df, x="paragraph_word_count", title= f"{len(files)} chapters stats")
fig
```

```python
fig.write_html("experiments/word_distribution.html")
```

```python
import pandas as pd
import plotly.express as px
import json
```

```python
data  = json.load(open("./output.json"))
data.keys()
```

```python
data = data["output"]
```

```python
merge_values = {}
for d in data:
    for k,v in d.items():
        if k in merge_values:
            merge_values[k].append(v)
        else:
            merge_values[k] = [v]

print(merge_values)
```

```python
df = pd.DataFrame(merge_values)
df
```

```python
df_output = df.mean(0).reset_index()
df_output.columns = ['(wd,th)', 'eval-windowdiff']
df_output
```
